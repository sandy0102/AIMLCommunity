```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
##our first model is Logistic Regression
def logreg (X_train, Y_train):
    from sklearn.linear_model import LogisticRegression
    log=LogisticRegression (random_state=0)
    log.fit(X_train, Y_train)
    print("Logistic Regression Training Accuracy:", log.score(X_train, Y_train))
    return log
```

```python
logrex=logreg(X_train, Y_train)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, logrex.predict(X_test))
print(cm)
```

```python
TP=cm[0][0]
TN=cm[1][1]
FN=cm[1][0]
FP=cm[0][1]
print("Testing accuracy of logistic regression model:", (TP+TN)/(TP+TN+FN+FP))
```

```python
# Our second model is Decision tree classifier
def logregDT (X_train, Y_train):
    from sklearn.tree import DecisionTreeClassifier
    #dtc = DecisionTreeClassifier(random_state=0)
    dtc = DecisionTreeClassifier(criterion="entropy", max_depth=3,random_state=0)
    dtc.fit(X_train, Y_train)
    print("Decision Tree Training Accuracy:{:.2f}%".format(dtc.score(X_train, Y_train)*100))
    return dtc
```

```python
logregDT=logregDT(X_train, Y_train)
```

```python
# Test the model
Y_predict = logregDT.predict(X_test)

# Evaluate the model
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_predict)
print(cm)
```

```python
TP=cm[0][0]
TN=cm[1][1]
FN=cm[1][0]
FP=cm[0][1]
print("Testing accuracy of Decision Tree model:", (TP+TN)/(TP+TN+FN+FP))
```

```python
##Our third model is Random Forest 
def logregRF(X_train, Y_train):
    from sklearn.ensemble import RandomForestClassifier
    rf = RandomForestClassifier(n_estimators= 1000, random_state = 0)
    rf.fit(X_train, Y_train)
    print("Random Forest Training Accuracy: {:.2f}%".format(rf.score(X_train, Y_train)*100))
    return rf
```

```python
logregRF=logregRF(X_train, Y_train)
```

```python
# Test the model
Y_predict = logregRF.predict(X_test)
```

```python
## Evaluate the model
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_predict)
print(cm)
```

```python
TP=cm[0][0]
TN=cm[1][1]
FN=cm[1][0]
FP=cm[0][1]
print("Random Forest Testing accuracy =", (TP+TN)/(TP+TN+FN+FP))
```
